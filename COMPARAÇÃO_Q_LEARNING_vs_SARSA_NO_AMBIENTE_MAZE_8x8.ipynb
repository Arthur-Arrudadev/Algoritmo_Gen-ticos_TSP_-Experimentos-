```python
# ============================================================================
# ATIVIDADE 10: COMPARA√á√ÉO Q-LEARNING vs SARSA NO AMBIENTE MAZE 8x8
# COM HISTOGRAMAS E VISUALIZA√á√ÉO DE PASSOS
#
# OBJETIVO: Comparar os algoritmos Q-Learning e SARSA no ambiente Maze 8x8
#           usando 100 epis√≥dios de treinamento e avalia√ß√£o
# ============================================================================

# Importa√ß√£o das bibliotecas necess√°rias
import numpy as np  # Para opera√ß√µes matem√°ticas e arrays
import matplotlib.pyplot as plt  # Para criar gr√°ficos e visualiza√ß√µes
import seaborn as sns  # Para estiliza√ß√£o dos gr√°ficos
from collections import defaultdict  # Para dicion√°rios com valores padr√£o
import random  # Para gerar n√∫meros aleat√≥rios (a√ß√µes explorat√≥rias)
import time  # Para pausas entre execu√ß√µes
from IPython.display import clear_output  # Para limpar sa√≠da em notebooks

# ============================================================================
# CONFIGURA√á√ïES DE ESTILO PARA OS GR√ÅFICOS
# ============================================================================

plt.style.use('seaborn-v0_8-darkgrid')  # Define estilo escuro com grade para gr√°ficos
sns.set_palette("husl")  # Define paleta de cores harmoniosa para visualiza√ß√µes

# ============================================================================
# IMPLEMENTA√á√ÉO DO AMBIENTE MAZE 8x8
# ============================================================================

class Maze:
    """
    Classe que representa o ambiente do labirinto 8x8.
    O agente come√ßa em (0,0) e deve chegar em (7,7) evitando buracos.
    """

    def __init__(self):
        """Inicializa o ambiente do labirinto com configura√ß√£o fixa."""
        self.size = 8  # Tamanho do labirinto: 8x8
        self.start = (0, 0)  # Posi√ß√£o inicial: canto superior esquerdo
        self.goal = (7, 7)   # Posi√ß√£o objetivo: canto inferior direito

        # Matriz 8x8 que define a localiza√ß√£o dos buracos (1) e caminhos livres (0)
        # 0 = caminho livre, 1 = buraco (armadilha)
        self.holes = np.array([
            [0, 0, 0, 0, 0, 1, 0, 0],  # Linha 0: buraco na coluna 5
            [0, 1, 1, 1, 0, 1, 0, 1],  # Linha 1: m√∫ltiplos buracos
            [0, 0, 0, 0, 0, 0, 0, 0],  # Linha 2: completamente livre
            [1, 1, 0, 1, 1, 1, 1, 0],  # Linha 3: quase todos buracos
            [0, 0, 0, 0, 0, 0, 0, 0],  # Linha 4: completamente livre
            [0, 1, 1, 1, 1, 1, 0, 1],  # Linha 5: muitos buracos
            [0, 0, 0, 0, 0, 0, 0, 0],  # Linha 6: completamente livre
            [0, 1, 0, 1, 1, 0, 1, 0]   # Linha 7: padr√£o alternado
        ])

        self.state = self.start  # Estado atual do agente (posi√ß√£o)
        self.actions = [0, 1, 2, 3]  # A√ß√µes poss√≠veis: 0=cima, 1=direita, 2=baixo, 3=esquerda

    def reset(self):
        """Reinicia o ambiente para o estado inicial."""
        self.state = self.start  # Retorna o agente para o in√≠cio
        return self.state  # Retorna o estado inicial

    def step(self, action):
        """
        Executa uma a√ß√£o no ambiente e retorna o resultado.

        Args:
            action: Inteiro representando a a√ß√£o (0-3)

        Returns:
            tuple: (novo_estado, recompensa, done)
            - novo_estado: nova posi√ß√£o (linha, coluna)
            - recompensa: valor da recompensa obtida
            - done: booleano indicando se o epis√≥dio terminou
        """
        row, col = self.state  # Desempacota a posi√ß√£o atual

        # Executa a a√ß√£o escolhida
        if action == 0:  # Mover para cima
            row = max(0, row - 1)  # Impede de sair do labirinto (n√£o pode ser negativo)
        elif action == 1:  # Mover para direita
            col = min(self.size - 1, col + 1)  # Impede de sair do labirinto (m√°ximo 7)
        elif action == 2:  # Mover para baixo
            row = min(self.size - 1, row + 1)  # Impede de sair do labirinto (m√°ximo 7)
        elif action == 3:  # Mover para esquerda
            col = max(0, col - 1)  # Impede de sair do labirinto (n√£o pode ser negativo)

        self.state = (row, col)  # Atualiza o estado com a nova posi√ß√£o

        # Verifica o resultado da a√ß√£o
        if self.holes[row, col] == 1:  # Caiu em um buraco
            done = True  # Epis√≥dio termina
            reward = -10  # Penalidade forte por cair em buraco
        elif self.state == self.goal:  # Chegou ao objetivo
            done = True  # Epis√≥dio termina
            reward = 10  # Recompensa m√°xima por alcan√ßar objetivo
        else:  # Movimento normal
            done = False  # Epis√≥dio continua
            reward = -0.1  # Pequena penalidade para incentivar caminhos curtos

        return self.state, reward, done

    def render(self):
        """Renderiza o labirinto com visualiza√ß√£o gr√°fica usando matplotlib."""
        fig, ax = plt.subplots(figsize=(8, 8))  # Cria figura 8x8 polegadas

        # Configura√ß√µes do gr√°fico
        ax.set_xlim(-0.5, 7.5)  # Limites do eixo X
        ax.set_ylim(-0.5, 7.5)  # Limites do eixo Y
        ax.set_aspect('equal')  # Mant√©m propor√ß√£o quadrada
        ax.set_xticks(range(8))  # Marcas no eixo X (0-7)
        ax.set_yticks(range(8))  # Marcas no eixo Y (0-7)
        ax.set_xticklabels(range(8))  # R√≥tulos do eixo X
        ax.set_yticklabels(range(8))  # R√≥tulos do eixo Y
        ax.grid(True, alpha=0.3)  # Adiciona grade com transpar√™ncia
        ax.set_title('Labirinto 8x8', fontsize=14, fontweight='bold')  # T√≠tulo
        ax.set_xlabel('Coluna')  # R√≥tulo do eixo X
        ax.set_ylabel('Linha')   # R√≥tulo do eixo Y

        # Desenha o labirinto c√©lula por c√©lula
        for i in range(8):  # Para cada linha
            for j in range(8):  # Para cada coluna
                if (i, j) == self.start:  # C√©lula de in√≠cio
                    rect = plt.Rectangle((j-0.5, 7-i-0.5), 1, 1,
                                       facecolor='lightgreen', edgecolor='black', linewidth=2)
                    ax.add_patch(rect)
                    ax.text(j, 7-i, 'IN√çCIO', ha='center', va='center', fontweight='bold')
                elif (i, j) == self.goal:  # C√©lula de objetivo
                    rect = plt.Rectangle((j-0.5, 7-i-0.5), 1, 1,
                                       facecolor='gold', edgecolor='black', linewidth=2)
                    ax.add_patch(rect)
                    ax.text(j, 7-i, 'OBJETIVO', ha='center', va='center', fontweight='bold')
                elif self.holes[i, j] == 1:  # C√©lula com buraco
                    rect = plt.Rectangle((j-0.5, 7-i-0.5), 1, 1,
                                       facecolor='lightcoral', edgecolor='black', linewidth=1)
                    ax.add_patch(rect)
                    ax.text(j, 7-i, 'X', ha='center', va='center', fontweight='bold', fontsize=12)
                else:  # C√©lula livre
                    rect = plt.Rectangle((j-0.5, 7-i-0.5), 1, 1,
                                       facecolor='white', edgecolor='black', linewidth=1)
                    ax.add_patch(rect)

        # Marca a posi√ß√£o atual do agente com um c√≠rculo azul
        row, col = self.state
        circle = plt.Circle((col, 7-row), 0.3, color='blue', alpha=0.7)
        ax.add_patch(circle)

        plt.tight_layout()  # Ajusta layout para evitar sobreposi√ß√£o
        plt.show()  # Mostra o gr√°fico

        self.render_text()  # Tamb√©m mostra vers√£o textual

    def render_text(self):
        """Mostra uma vers√£o textual simplificada do labirinto no console."""
        print("\n" + "=" * 33)
        print("LABIRINTO 8x8 - Vista Textual")
        print("=" * 33)
        print("Legenda: S=In√≠cio, G=Objetivo, X=Buraco, O=Agente")
        print()

        for i in range(8):
            linha = ""
            for j in range(8):
                if (i, j) == self.start and (i, j) == self.state:
                    linha += "[O] "  # Agente no in√≠cio
                elif (i, j) == self.start:
                    linha += "[S] "  # In√≠cio
                elif (i, j) == self.goal and (i, j) == self.state:
                    linha += "[O] "  # Agente no objetivo
                elif (i, j) == self.goal:
                    linha += "[G] "  # Objetivo
                elif (i, j) == self.state:
                    linha += "[O] "  # Agente
                elif self.holes[i, j] == 1:
                    linha += "[X] "  # Buraco
                else:
                    linha += "[ ] "  # Caminho livre
            print(linha)

        print(f"\nPosi√ß√£o atual: {self.state}")
        print("=" * 33)

# ============================================================================
# IMPLEMENTA√á√ÉO DO AGENTE Q-LEARNING
# ============================================================================

class QLearningAgent:
    """
    Agente que implementa o algoritmo Q-Learning (off-policy).
    Aprende da pol√≠tica √≥tima usando a equa√ß√£o:
    Q(s,a) ‚Üê Q(s,a) + Œ±[r + Œ≥*maxQ(s',a') - Q(s,a)]
    """

    def __init__(self, env, alpha=0.1, gamma=0.9, epsilon=0.1, epsilon_decay=0.995):
        """
        Inicializa o agente Q-Learning com par√¢metros de aprendizado.

        Args:
            env: Ambiente Maze
            alpha: Taxa de aprendizado (0.1) - controla quanto aprende de cada experi√™ncia
            gamma: Fator de desconto (0.9) - import√¢ncia das recompensas futuras
            epsilon: Taxa de explora√ß√£o inicial (0.1) - probabilidade de a√ß√£o aleat√≥ria
            epsilon_decay: Taxa de decaimento da explora√ß√£o (0.995) - reduz epsilon ao longo do tempo
        """
        self.env = env  # Refer√™ncia ao ambiente
        self.alpha = alpha  # Taxa de aprendizado
        self.gamma = gamma  # Fator de desconto
        self.epsilon = epsilon  # Taxa de explora√ß√£o inicial
        self.epsilon_decay = epsilon_decay  # Taxa de decaimento da explora√ß√£o
        self.epsilon_min = 0.01  # Valor m√≠nimo de epsilon

        # Tabela Q: mapeia estados para valores de cada a√ß√£o
        # defaultdict cria automaticamente arrays de zeros para novos estados
        self.q_table = defaultdict(lambda: np.zeros(len(env.actions)))

        # Hist√≥ricos para an√°lise
        self.episode_rewards = []  # Recompensas totais por epis√≥dio
        self.episode_lengths = []  # N√∫mero de passos por epis√≥dio
        self.epsilon_history = []  # Hist√≥rico de valores de epsilon
        self.all_rewards = []  # Todas as recompensas (para histogramas)

    def choose_action(self, state):
        """
        Escolhe uma a√ß√£o usando pol√≠tica Œµ-greedy.

        Args:
            state: Estado atual (posi√ß√£o no labirinto)

        Returns:
            int: A√ß√£o escolhida (0-3)
        """
        if random.random() < self.epsilon:
            # Explora√ß√£o: escolhe a√ß√£o aleat√≥ria com probabilidade Œµ
            return random.choice(self.env.actions)
        else:
            # Explora√ß√£o: escolhe a√ß√£o com maior valor Q
            return np.argmax(self.q_table[state])

    def update(self, state, action, reward, next_state, done):
        """
        Atualiza a tabela Q usando a equa√ß√£o do Q-Learning.

        Args:
            state: Estado atual
            action: A√ß√£o executada
            reward: Recompensa obtida
            next_state: Pr√≥ximo estado
            done: Se o epis√≥dio terminou
        """
        current_q = self.q_table[state][action]  # Valor Q atual

        # Calcula o target (valor alvo)
        if done:
            target = reward  # Se terminou, target √© apenas a recompensa
        else:
            # Q-Learning: usa o m√°ximo valor Q do pr√≥ximo estado (pol√≠tica √≥tima)
            target = reward + self.gamma * np.max(self.q_table[next_state])

        # Atualiza√ß√£o da tabela Q usando a f√≥rmula do Q-Learning
        self.q_table[state][action] += self.alpha * (target - current_q)

    def train(self, episodes=100, max_steps=200):
        """
        Treina o agente por um n√∫mero espec√≠fico de epis√≥dios.

        Args:
            episodes: N√∫mero de epis√≥dios de treinamento (100)
            max_steps: M√°ximo de passos por epis√≥dio (200)

        Returns:
            dict: Tabela Q aprendida
        """
        print("üöÄ TREINAMENTO Q-LEARNING")
        print(f"Par√¢metros: Œ±={self.alpha}, Œ≥={self.gamma}, Œµ={self.epsilon}")
        print(f"Epis√≥dios: {episodes}")
        print("-" * 50)

        for episode in range(episodes):
            state = self.env.reset()  # Reinicia ambiente para novo epis√≥dio
            total_reward = 0  # Acumula recompensa do epis√≥dio
            steps = 0  # Conta passos do epis√≥dio

            for step in range(max_steps):
                # Escolhe e executa a√ß√£o
                action = self.choose_action(state)
                next_state, reward, done = self.env.step(action)

                # Atualiza tabela Q com experi√™ncia (s, a, r, s')
                self.update(state, action, reward, next_state, done)

                # Prepara para pr√≥ximo passo
                state = next_state
                total_reward += reward
                steps += 1

                if done:  # Se epis√≥dio terminou, sai do loop
                    break

            # Decai epsilon (reduz explora√ß√£o ao longo do tempo)
            self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

            # Armazena estat√≠sticas do epis√≥dio
            self.episode_rewards.append(total_reward)
            self.episode_lengths.append(steps)
            self.epsilon_history.append(self.epsilon)
            self.all_rewards.append(total_reward)  # Para histogramas

            # Feedback peri√≥dico a cada 100 epis√≥dios
            if (episode + 1) % 100 == 0:
                avg_reward = np.mean(self.episode_rewards[-100:])
                avg_steps = np.mean(self.episode_lengths[-100:])
                print(f"Epis√≥dio {episode + 1}: Recompensa = {avg_reward:.2f}, "
                      f"Passos = {avg_steps:.1f}, Œµ = {self.epsilon:.3f}")

        print("‚úÖ Q-LEARNING CONCLU√çDO!")
        return self.q_table

    def evaluate(self, episodes=100, max_steps=200):
        """
        Avalia o agente treinado sem explora√ß√£o (epsilon = 0).

        Args:
            episodes: N√∫mero de epis√≥dios de avalia√ß√£o (100)
            max_steps: M√°ximo de passos por epis√≥dio (200)

        Returns:
            tuple: (taxa_sucesso, recompensa_m√©dia, passos_m√©dios, recompensas, passos)
        """
        original_epsilon = self.epsilon  # Salva valor original
        self.epsilon = 0  # Desativa explora√ß√£o (s√≥ escolhe melhor a√ß√£o)

        rewards = []  # Recompensas de cada epis√≥dio de avalia√ß√£o
        steps_list = []  # Passos de cada epis√≥dio
        successes = 0  # Contador de sucessos (chegou ao objetivo)

        for episode in range(episodes):
            state = self.env.reset()
            total_reward = 0
            steps = 0

            for step in range(max_steps):
                action = self.choose_action(state)  # Sempre escolhe melhor a√ß√£o
                next_state, reward, done = self.env.step(action)

                state = next_state
                total_reward += reward
                steps += 1

                if done:
                    if state == self.env.goal:
                        successes += 1  # Conta como sucesso se chegou ao objetivo
                    break

            rewards.append(total_reward)
            steps_list.append(steps)

        self.epsilon = original_epsilon  # Restaura epsilon original

        # Calcula estat√≠sticas
        success_rate = successes / episodes * 100  # Taxa de sucesso em porcentagem
        avg_reward = np.mean(rewards)  # Recompensa m√©dia
        avg_steps = np.mean(steps_list)  # Passos m√©dios

        return success_rate, avg_reward, avg_steps, rewards, steps_list

# ============================================================================
# IMPLEMENTA√á√ÉO DO AGENTE SARSA
# ============================================================================

class SarsaAgent:
    """
    Agente que implementa o algoritmo SARSA (on-policy).
    Aprende da pol√≠tica atual usando a equa√ß√£o:
    Q(s,a) ‚Üê Q(s,a) + Œ±[r + Œ≥*Q(s',a') - Q(s,a)]
    """

    def __init__(self, env, alpha=0.1, gamma=0.9, epsilon=0.1, epsilon_decay=0.995):
        """
        Inicializa o agente SARSA com par√¢metros id√™nticos ao Q-Learning
        para compara√ß√£o justa.
        """
        self.env = env
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = 0.01

        self.q_table = defaultdict(lambda: np.zeros(len(env.actions)))
        self.episode_rewards = []
        self.episode_lengths = []
        self.epsilon_history = []
        self.all_rewards = []

    def choose_action(self, state):
        """Mesma pol√≠tica Œµ-greedy do Q-Learning."""
        if random.random() < self.epsilon:
            return random.choice(self.env.actions)
        else:
            return np.argmax(self.q_table[state])

    def update(self, state, action, reward, next_state, next_action, done):
        """
        Atualiza a tabela Q usando a equa√ß√£o do SARSA.

        Diferen√ßa principal do Q-Learning: usa Q(s',a') em vez de maxQ(s',a')
        (aprende da pol√≠tica atual, n√£o da √≥tima).
        """
        current_q = self.q_table[state][action]

        if done:
            target = reward
        else:
            # SARSA: usa o valor Q da pr√≥xima a√ß√£o (n√£o o m√°ximo)
            target = reward + self.gamma * self.q_table[next_state][next_action]

        self.q_table[state][action] += self.alpha * (target - current_q)

    def train(self, episodes=100, max_steps=200):
        """
        Treina o agente SARSA.

        Diferen√ßa do Q-Learning: precisa escolher pr√≥xima a√ß√£o antes de atualizar.
        """
        print("üöÄ TREINAMENTO SARSA")
        print(f"Par√¢metros: Œ±={self.alpha}, Œ≥={self.gamma}, Œµ={self.epsilon}")
        print(f"Epis√≥dios: {episodes}")
        print("-" * 50)

        for episode in range(episodes):
            state = self.env.reset()
            action = self.choose_action(state)  # Escolhe primeira a√ß√£o

            total_reward = 0
            steps = 0

            for step in range(max_steps):
                # Executa a√ß√£o
                next_state, reward, done = self.env.step(action)

                # Escolhe pr√≥xima a√ß√£o (necess√°rio para SARSA)
                next_action = self.choose_action(next_state)

                # Atualiza tabela Q usando SARSA (precisa da pr√≥xima a√ß√£o)
                self.update(state, action, reward, next_state, next_action, done)

                # Prepara para pr√≥ximo passo
                state = next_state
                action = next_action
                total_reward += reward
                steps += 1

                if done:
                    break

            self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

            self.episode_rewards.append(total_reward)
            self.episode_lengths.append(steps)
            self.epsilon_history.append(self.epsilon)
            self.all_rewards.append(total_reward)

            if (episode + 1) % 100 == 0:
                avg_reward = np.mean(self.episode_rewards[-100:])
                avg_steps = np.mean(self.episode_lengths[-100:])
                print(f"Epis√≥dio {episode + 1}: Recompensa = {avg_reward:.2f}, "
                      f"Passos = {avg_steps:.1f}, Œµ = {self.epsilon:.3f}")

        print("‚úÖ SARSA CONCLU√çDO!")
        return self.q_table

    def evaluate(self, episodes=100, max_steps=200):
        """Avalia√ß√£o similar ao Q-Learning para compara√ß√£o justa."""
        original_epsilon = self.epsilon
        self.epsilon = 0

        rewards = []
        steps_list = []
        successes = 0

        for episode in range(episodes):
            state = self.env.reset()
            total_reward = 0
            steps = 0

            for step in range(max_steps):
                action = self.choose_action(state)
                next_state, reward, done = self.env.step(action)

                state = next_state
                total_reward += reward
                steps += 1

                if done:
                    if state == self.env.goal:
                        successes += 1
                    break

            rewards.append(total_reward)
            steps_list.append(steps)

        self.epsilon = original_epsilon

        success_rate = successes / episodes * 100
        avg_reward = np.mean(rewards)
        avg_steps = np.mean(steps_list)

        return success_rate, avg_reward, avg_steps, rewards, steps_list

# ============================================================================
# FUN√á√ïES PARA VISUALIZA√á√ÉO DE RESULTADOS
# ============================================================================

def plot_histograms(q_rewards, sarsa_rewards, q_steps, sarsa_steps):
    """
    Cria 4 histogramas comparativos entre Q-Learning e SARSA.

    Args:
        q_rewards: Lista de recompensas do Q-Learning
        sarsa_rewards: Lista de recompensas do SARSA
        q_steps: Lista de passos do Q-Learning
        sarsa_steps: Lista de passos do SARSA
    """
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid de gr√°ficos

    # Histograma 1: Distribui√ß√£o de Recompensas - Q-Learning
    ax1 = axes[0, 0]
    ax1.hist(q_rewards, bins=20, alpha=0.7, color='blue', edgecolor='black')
    ax1.set_xlabel('Recompensa')
    ax1.set_ylabel('Frequ√™ncia')
    ax1.set_title('Distribui√ß√£o de Recompensas - Q-Learning')
    ax1.grid(True, alpha=0.3)
    ax1.axvline(np.mean(q_rewards), color='red', linestyle='--',
                label=f'M√©dia: {np.mean(q_rewards):.2f}')
    ax1.legend()

    # Histograma 2: Distribui√ß√£o de Recompensas - SARSA
    ax2 = axes[0, 1]
    ax2.hist(sarsa_rewards, bins=20, alpha=0.7, color='red', edgecolor='black')
    ax2.set_xlabel('Recompensa')
    ax2.set_ylabel('Frequ√™ncia')
    ax2.set_title('Distribui√ß√£o de Recompensas - SARSA')
    ax2.grid(True, alpha=0.3)
    ax2.axvline(np.mean(sarsa_rewards), color='blue', linestyle='--',
                label=f'M√©dia: {np.mean(sarsa_rewards):.2f}')
    ax2.legend()

    # Histograma 3: Distribui√ß√£o de Passos - Q-Learning
    ax3 = axes[1, 0]
    ax3.hist(q_steps, bins=20, alpha=0.7, color='blue', edgecolor='black')
    ax3.set_xlabel('Passos por Epis√≥dio')
    ax3.set_ylabel('Frequ√™ncia')
    ax3.set_title('Distribui√ß√£o de Passos - Q-Learning')
    ax3.grid(True, alpha=0.3)
    ax3.axvline(np.mean(q_steps), color='red', linestyle='--',
                label=f'M√©dia: {np.mean(q_steps):.1f}')
    ax3.legend()

    # Histograma 4: Distribui√ß√£o de Passos - SARSA
    ax4 = axes[1, 1]
    ax4.hist(sarsa_steps, bins=20, alpha=0.7, color='red', edgecolor='black')
    ax4.set_xlabel('Passos por Epis√≥dio')
    ax4.set_ylabel('Frequ√™ncia')
    ax4.set_title('Distribui√ß√£o de Passos - SARSA')
    ax4.grid(True, alpha=0.3)
    ax4.axvline(np.mean(sarsa_steps), color='blue', linestyle='--',
                label=f'M√©dia: {np.mean(sarsa_steps):.1f}')
    ax4.legend()

    plt.suptitle('COMPARA√á√ÉO DE DISTRIBUI√á√ïES: Q-LEARNING vs SARSA',
                fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

def plot_step_analysis(q_agent, sarsa_agent):
    """
    An√°lise detalhada dos passos realizados durante o treinamento.

    Args:
        q_agent: Agente Q-Learning treinado
        sarsa_agent: Agente SARSA treinado
    """
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # Gr√°fico 1: Evolu√ß√£o dos passos durante treinamento
    ax1 = axes[0, 0]

    window = 50  # Tamanho da janela para m√©dia m√≥vel
    q_steps_ma = np.convolve(q_agent.episode_lengths, np.ones(window)/window, mode='valid')
    sarsa_steps_ma = np.convolve(sarsa_agent.episode_lengths, np.ones(window)/window, mode='valid')

    ax1.plot(q_steps_ma, label='Q-Learning', color='blue', alpha=0.7, linewidth=2)
    ax1.plot(sarsa_steps_ma, label='SARSA', color='red', alpha=0.7, linewidth=2)
    ax1.set_xlabel('Epis√≥dio')
    ax1.set_ylabel('Passos (M√©dia M√≥vel)')
    ax1.set_title('Evolu√ß√£o dos Passos Durante Treinamento')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Gr√°fico 2: Boxplot comparativo de passos
    ax2 = axes[0, 1]

    data = [q_agent.episode_lengths, sarsa_agent.episode_lengths]
    labels = ['Q-Learning', 'SARSA']

    bp = ax2.boxplot(data, labels=labels, patch_artist=True)
    bp['boxes'][0].set_facecolor('blue')
    bp['boxes'][1].set_facecolor('red')

    ax2.set_ylabel('Passos por Epis√≥dio')
    ax2.set_title('Distribui√ß√£o dos Passos (Boxplot)')
    ax2.grid(True, alpha=0.3)

    # Gr√°fico 3: Taxa de sucesso por faixa de passos
    ax3 = axes[1, 0]

    bins = [0, 10, 20, 30, 40, 50, 100, 200]  # Faixas de passos

    q_steps_hist, _ = np.histogram(q_agent.episode_lengths, bins=bins)
    sarsa_steps_hist, _ = np.histogram(sarsa_agent.episode_lengths, bins=bins)

    x = np.arange(len(bins)-1)
    width = 0.35

    ax3.bar(x - width/2, q_steps_hist, width, label='Q-Learning', color='blue', alpha=0.7)
    ax3.bar(x + width/2, sarsa_steps_hist, width, label='SARSA', color='red', alpha=0.7)

    ax3.set_xlabel('Faixa de Passos')
    ax3.set_ylabel('Frequ√™ncia')
    ax3.set_title('Distribui√ß√£o por Faixa de Passos')
    ax3.set_xticks(x)
    ax3.set_xticklabels([f'{bins[i]}-{bins[i+1]}' for i in range(len(bins)-1)], rotation=45)
    ax3.legend()
    ax3.grid(True, alpha=0.3)

    # Gr√°fico 4: Correla√ß√£o entre recompensa e passos
    ax4 = axes[1, 1]

    # Amostra aleat√≥ria para evitar sobrecarga no gr√°fico
    sample_size = min(200, len(q_agent.episode_lengths))
    indices = np.random.choice(len(q_agent.episode_lengths), sample_size, replace=False)

    q_lengths_sample = np.array(q_agent.episode_lengths)[indices]
    q_rewards_sample = np.array(q_agent.episode_rewards)[indices]
    sarsa_lengths_sample = np.array(sarsa_agent.episode_lengths)[indices]
    sarsa_rewards_sample = np.array(sarsa_agent.episode_rewards)[indices]

    ax4.scatter(q_lengths_sample, q_rewards_sample, alpha=0.5, color='blue',
                label='Q-Learning', s=30)
    ax4.scatter(sarsa_lengths_sample, sarsa_rewards_sample, alpha=0.5, color='red',
                label='SARSA', s=30)

    ax4.set_xlabel('Passos')
    ax4.set_ylabel('Recompensa')
    ax4.set_title('Correla√ß√£o: Passos vs Recompensa')
    ax4.legend()
    ax4.grid(True, alpha=0.3)

    plt.suptitle('AN√ÅLISE DETALHADA DOS PASSOS REALIZADOS',
                fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# ============================================================================
# EXPERIMENTO PRINCIPAL - EXECUTA A COMPARA√á√ÉO COMPLETA
# ============================================================================

def executar_experimento_completo():
    """
    Executa o experimento completo de compara√ß√£o entre Q-Learning e SARSA.
    Divide em 5 fases: treinamento, avalia√ß√£o, visualiza√ß√£o, an√°lise e demonstra√ß√£o.

    Returns:
        dict: Resultados completos do experimento
    """

    print("=" * 70)
    print("üéØ EXPERIMENTO COMPLETO: Q-LEARNING vs SARSA")
    print("=" * 70)

    # ========================================================================
    # FASE 0: DEFINI√á√ÉO DOS PAR√ÇMETROS
    # ========================================================================

    parametros = {
        'episodes_train': 100,      # 100 epis√≥dios de treinamento
        'episodes_eval': 100,       # 100 epis√≥dios de avalia√ß√£o
        'max_steps': 200,           # M√°ximo 200 passos por epis√≥dio
        'alpha': 0.1,               # Taxa de aprendizado
        'gamma': 0.9,               # Fator de desconto
        'epsilon': 0.1,             # Taxa de explora√ß√£o inicial
        'epsilon_decay': 0.995      # Decaimento da explora√ß√£o
    }

    print("\nüìã PAR√ÇMETROS DO EXPERIMENTO")
    print("-" * 40)
    for param, valor in parametros.items():
        print(f"{param}: {valor}")

    # ========================================================================
    # FASE 1: TREINAMENTO DOS AGENTES
    # ========================================================================
    print("\n" + "=" * 70)
    print("üèãÔ∏è FASE 1: TREINAMENTO")
    print("=" * 70)

    # Treina agente Q-Learning
    print("\nüîµ TREINANDO Q-LEARNING...")
    q_agent = QLearningAgent(
        env=Maze(),
        alpha=parametros['alpha'],
        gamma=parametros['gamma'],
        epsilon=parametros['epsilon'],
        epsilon_decay=parametros['epsilon_decay']
    )
    q_agent.train(
        episodes=parametros['episodes_train'],
        max_steps=parametros['max_steps']
    )

    # Treina agente SARSA
    print("\nüî¥ TREINANDO SARSA...")
    sarsa_agent = SarsaAgent(
        env=Maze(),
        alpha=parametros['alpha'],
        gamma=parametros['gamma'],
        epsilon=parametros['epsilon'],
        epsilon_decay=parametros['epsilon_decay']
    )
    sarsa_agent.train(
        episodes=parametros['episodes_train'],
        max_steps=parametros['max_steps']
    )

    # ========================================================================
    # FASE 2: AVALIA√á√ÉO DOS AGENTES TREINADOS
    # ========================================================================
    print("\n" + "=" * 70)
    print("üìä FASE 2: AVALIA√á√ÉO")
    print("=" * 70)

    # Avalia Q-Learning
    print("\nüîµ AVALIANDO Q-LEARNING...")
    q_success, q_reward, q_steps, q_eval_rewards, q_eval_steps = q_agent.evaluate(
        episodes=parametros['episodes_eval'],
        max_steps=parametros['max_steps']
    )

    # Avalia SARSA
    print("\nüî¥ AVALIANDO SARSA...")
    sarsa_success, sarsa_reward, sarsa_steps, sarsa_eval_rewards, sarsa_eval_steps = sarsa_agent.evaluate(
        episodes=parametros['episodes_eval'],
        max_steps=parametros['max_steps']
    )

    # ========================================================================
    # FASE 3: VISUALIZA√á√ÉO DOS RESULTADOS
    # ========================================================================
    print("\n" + "=" * 70)
    print("üìà FASE 3: VISUALIZA√á√ÉO E AN√ÅLISE")
    print("=" * 70)

    # 3.1 Histogramas comparativos
    print("\nüìä HISTOGRAMAS...")
    plot_histograms(q_eval_rewards, sarsa_eval_rewards, q_eval_steps, sarsa_eval_steps)

    # 3.2 An√°lise detalhada dos passos
    print("\nüìà ANALISANDO PASSOS REALIZADOS...")
    plot_step_analysis(q_agent, sarsa_agent)

    # 3.3 Gr√°fico comparativo principal
    print("\nüé® CRIANDO GR√ÅFICO COMPARATIVO PRINCIPAL...")

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Gr√°fico 1: Compara√ß√£o de desempenho (barras)
    ax1 = axes[0]

    metrics = ['Taxa Sucesso', 'Recompensa', 'Passos']
    q_values = [q_success, q_reward, q_steps]
    sarsa_values = [sarsa_success, sarsa_reward, sarsa_steps]

    x = np.arange(len(metrics))
    width = 0.35

    bars1 = ax1.bar(x - width/2, q_values, width, label='Q-Learning', color='blue', alpha=0.7)
    bars2 = ax1.bar(x + width/2, sarsa_values, width, label='SARSA', color='red', alpha=0.7)

    ax1.set_xlabel('M√©trica')
    ax1.set_ylabel('Valor')
    ax1.set_title('Compara√ß√£o de Desempenho')
    ax1.set_xticks(x)
    ax1.set_xticklabels(metrics)
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Adiciona valores nas barras
    for bars in [bars1, bars2]:
        for bar in bars:
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                    f'{height:.1f}', ha='center', va='bottom')

    # Gr√°fico 2: Curvas de aprendizado (linhas)
    ax2 = axes[1]

    window = 50
    q_rewards_ma = np.convolve(q_agent.episode_rewards, np.ones(window)/window, mode='valid')
    sarsa_rewards_ma = np.convolve(sarsa_agent.episode_rewards, np.ones(window)/window, mode='valid')

    ax2.plot(q_rewards_ma, label='Q-Learning', color='blue', alpha=0.7, linewidth=2)
    ax2.plot(sarsa_rewards_ma, label='SARSA', color='red', alpha=0.7, linewidth=2)

    ax2.set_xlabel('Epis√≥dio')
    ax2.set_ylabel('Recompensa M√©dia')
    ax2.set_title('Curvas de Aprendizado (M√©dia M√≥vel)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.suptitle('COMPARA√á√ÉO COMPLETA: Q-LEARNING vs SARSA',
                fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # ========================================================================
    # FASE 4: AN√ÅLISE ESTAT√çSTICA E CONCLUS√ïES
    # ========================================================================
    print("\n" + "=" * 70)
    print("üìù FASE 4: RESULTADOS E CONCLUS√ïES")
    print("=" * 70)

    # Tabela comparativa formatada
    print(f"\nüìä RESULTADOS DA AVALIA√á√ÉO ({parametros['episodes_eval']} epis√≥dios):")
    print("-" * 65)
    print(f"{'M√©trica':<20} {'Q-Learning':<15} {'SARSA':<15} {'Diferen√ßa':<15}")
    print("-" * 65)
    print(f"{'Taxa de Sucesso':<20} {q_success:<15.1f}% {sarsa_success:<15.1f}% {q_success-sarsa_success:<+15.1f}%")
    print(f"{'Recompensa M√©dia':<20} {q_reward:<15.2f} {sarsa_reward:<15.2f} {q_reward-sarsa_reward:<+15.2f}")
    print(f"{'Passos M√©dios':<20} {q_steps:<15.1f} {sarsa_steps:<15.1f} {q_steps-sarsa_steps:<+15.1f}")
    print("-" * 65)

    # An√°lise estat√≠stica detalhada
    print("\nüìà AN√ÅLISE ESTAT√çSTICA:")

    # Vari√¢ncia das recompensas (mede consist√™ncia)
    q_var = np.var(q_eval_rewards)
    sarsa_var = np.var(sarsa_eval_rewards)
    print(f"Vari√¢ncia das recompensas:")
    print(f"  Q-Learning: {q_var:.2f} (maior = menos consistente)")
    print(f"  SARSA: {sarsa_var:.2f} (menor = mais consistente)")

    # Estat√≠sticas descritivas dos passos
    print(f"\nEstat√≠sticas dos passos:")
    print(f"Q-Learning: M√©dia={np.mean(q_eval_steps):.1f}, "
          f"Mediana={np.median(q_eval_steps):.1f}, "
          f"Desvio={np.std(q_eval_steps):.1f}")
    print(f"SARSA: M√©dia={np.mean(sarsa_eval_steps):.1f}, "
          f"Mediana={np.median(sarsa_eval_steps):.1f}, "
          f"Desvio={np.std(sarsa_eval_steps):.1f}")

    # Conclus√£o baseada na taxa de sucesso
    if q_success > sarsa_success:
        melhor = "Q-Learning"
        razao = "busca caminhos mais curtos"
    else:
        melhor = "SARSA"
        razao = "√© mais conservador e evita buracos"

    print(f"\nüèÜ CONCLUS√ÉO: {melhor} apresentou melhor desempenho geral!")
    print(f"   Provavelmente porque {razao}.")

    # ========================================================================
    # FASE 5: DEMONSTRA√á√ÉO PR√ÅTICA DO MELHOR AGENTE
    # ========================================================================
    print("\n" + "=" * 70)
    print("üéÆ FASE 5: DEMONSTRA√á√ÉO DO MELHOR AGENTE")
    print("=" * 70)

    # Seleciona melhor agente baseado na taxa de sucesso
    if q_success >= sarsa_success:
        melhor_agente = q_agent
        nome_agente = "Q-Learning"
    else:
        melhor_agente = sarsa_agent
        nome_agente = "SARSA"

    print(f"\nDemonstrando {nome_agente}...")

    # Executa demonstra√ß√£o passo a passo
    demo_env = Maze()
    state = demo_env.reset()
    done = False
    steps = 0
    total_reward = 0
    path = [state]  # Armazena caminho percorrido

    print(f"\nIniciando no estado: {state}")
    demo_env.render()

    while not done and steps < 50:
        action = np.argmax(melhor_agente.q_table[state])  # Escolhe melhor a√ß√£o
        action_names = {0: "CIMA", 1: "DIREITA", 2: "BAIXO", 3: "ESQUERDA"}

        print(f"\nPasso {steps + 1}:")
        print(f"A√ß√£o escolhida: {action_names[action]}")

        next_state, reward, done = demo_env.step(action)

        path.append(next_state)
        total_reward += reward
        steps += 1

        print(f"Novo estado: {next_state}")
        print(f"Recompensa deste passo: {reward:.1f}")
        print(f"Recompensa acumulada: {total_reward:.1f}")

        demo_env.render()

        state = next_state

        if done:
            if state == demo_env.goal:
                print("\nüéâ OBJETIVO ALCAN√áADO COM SUCESSO!")
            else:
                print("\nüí• CAIU EM UM BURACO!")
        elif steps >= 50:
            print("\n‚è∞ LIMITE DE PASSOS ATINGIDO!")

        time.sleep(0.5)  # Pausa para visualiza√ß√£o

    # Resumo final da demonstra√ß√£o
    print(f"\nüìã RESUMO DA DEMONSTRA√á√ÉO:")
    print(f"Algoritmo: {nome_agente}")
    print(f"Passos realizados: {steps}")
    print(f"Recompensa final: {total_reward:.1f}")
    print(f"Caminho percorrido: {path}")

    # Retorna todos os resultados para poss√≠vel an√°lise posterior
    return {
        'q_agent': q_agent,
        'sarsa_agent': sarsa_agent,
        'q_results': (q_success, q_reward, q_steps, q_eval_rewards, q_eval_steps),
        'sarsa_results': (sarsa_success, sarsa_reward, sarsa_steps, sarsa_eval_rewards, sarsa_eval_steps)
    }

# ============================================================================
# EXECU√á√ÉO PRINCIPAL DO PROGRAMA
# ============================================================================

if __name__ == "__main__":
    """
    Ponto de entrada principal do programa.
    Executa o experimento completo quando o arquivo √© rodado diretamente.
    """
    print("=" * 70)
    print("üéÆ ATIVIDADE 10: COMPARA√á√ÉO Q-LEARNING vs SARSA")
    print("=" * 70)
    print("Discente: Arthur Arruda Teixeira")
    print("Docente: Rodrigo Lira")
    print("Disciplina: TEIA")
    print("=" * 70)

    # Executa o experimento completo
    try:
        resultados = executar_experimento_completo()
        print("\n" + "=" * 70)
        print("‚úÖ EXPERIMENTO CONCLU√çDO COM SUCESSO!")
        print("=" * 70)
    except Exception as e:
        # Tratamento de erros para debugging
        print(f"\n‚ùå ERRO DURANTE A EXECU√á√ÉO: {e}")
        print("=" * 70)



```
